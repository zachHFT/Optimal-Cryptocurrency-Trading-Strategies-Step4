# Web Scraping Medium articles with tags ML/DL/AI using BS & SELENIUM

```python
#import all libraries
import urllib.request
import requests
import bs4
from bs4 import BeautifulSoup
```

```python
#save the links we wanna scrape

kyber = 'https://medium.com/@kyberteam'
arificial_intelligence = "https://medium.com/tag/artificial-intelligence"
machine_learning = "https://medium.com/tag/machine-learning"
deep_learning = "https://medium.com/tag/deep-learning"
```

```python
from selenium import webdriver
```

### Dynamic loading with selenium

```python
driver = webdriver.Edge('/Users/liuye/Desktop/Everything/Driver_Notes/msedgedriver.exe')
driver.get(kyber)
```

### Load the webpage that is displayed till you reach the end to collect all tagged articles

```python
res = driver.execute_script("return document.documentElement.outerHTML")
driver.quit()
```

```python
soup = BeautifulSoup(res, 'lxml')
```

```python
t= soup.find_all('a', class_='eh bw')
title = []
for p in t:
    title.append(p.text)
```

```python
title
```

```python
a = soup.find_all('p', class_='ba b bb bc by')
date = []
for p in a:
    date.append(p.text)
date.pop(0)
date.pop(-1)
```

```python
date
```

```python
import pandas as pd
data = pd.DataFrame({'Date':date,'Title':title})
```

```python

data.to_csv('kybermedium.csv')
```
